# Data Preprocessing

```python
features = df.drop('letter', axis = 1)
target = df['letter']
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
le = LabelEncoder()
target = le.fit_transform(target)
```

```python
scaler = StandardScaler()
features = pd.DataFrame(scaler.fit_transform(features), columns = features.columns)
```

```python
X = features
y = target
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
ycat = to_categorical(y)
```

```python
ycat
```

Target labels are encoded; features are standardized.

## Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, ycat, test_size = 0.2, random_state = 42, stratify = ycat)
```

# Model Building

```python
from sklearn.model_selection import  GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
#!pip install scikeras
import scikeras
```

## Define model

Define a function for looping the model

```python
from scikeras.wrappers import KerasClassifier
```

```python
def create_model():
  model = Sequential()
  model.add(Dense(64, input_dim = X.shape[1], kernel_initializer='normal', activation = 'relu'))
  model.add(Dense(32, input_dim = X.shape[1], kernel_initializer = 'normal', activation = 'relu'))
  model.add(Dense(26, kernel_initializer = 'normal', activation = 'softmax'))

  adam = Adam(learning_rate = 0.001)
  model.compile(loss = 'categorical_crossentropy', optimizer = adam, metrics = ['accuracy'])
  return model
```

### Tuning Hyperparameters: Batch Size and Epochs

```python
# Early stop the execution if the output doesnt change for 5 consecutive attempts
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights= True)
```

```python
model = KerasClassifier(build_fn = create_model, verbose = 0, callbacks= [early_stop])
```

```python
param_grid = {
    'batch_size' :[8,16],
    'epochs' : [10,50]
}
```

```python
grid_search = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(),verbose = 0)
```

```python
#!pip uninstall -y scikit-learn
#!pip install scikit-learn==1.5.2
```

```python
grid_result = grid_search.fit(X_train,y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

### Tuning Hyperparameters: Learning rate and Drop out rate

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def create_model(learning_rate=0.001, dropout_rate=0.05):
  model = Sequential()
  model.add(Dense(64, input_dim = 16, kernel_initializer='normal', activation = 'relu'))
  model.add(Dropout(dropout_rate))
  model.add(Dense(32, input_dim = 16, kernel_initializer= 'normal', activation = 'relu'))
  model.add(Dropout(dropout_rate))
  model.add(Dense(26, kernel_initializer = 'normal', activation = 'softmax'))

  adam = Adam(learning_rate = learning_rate)
  model.compile(loss = 'categorical_crossentropy', optimizer = adam, metrics = ['accuracy'])
  return model
```

```python
model = KerasClassifier(build_fn = create_model, epochs = 20, batch_size = 8, verbose = 0, callbacks =[early_stop])
```

```python
param_dist = {
    'model__learning_rate': [0.001,0.003,0.01],
    'model__dropout_rate': [0.05,0.1]
}
```

```python
grid_search1 = RandomizedSearchCV(estimator = model, param_distributions = param_dist,
                                  cv = KFold(), verbose = 0, n_iter = 10, n_jobs = -1 )
```

```python
grid_result1 = grid_search1.fit(X_train, y_train)
```

```python
print("best accuracy %f using %s" % (grid_result1.best_score_, grid_result1.best_params_))
```

### Tuning Hyperparameters: Activation function and Kernel Initializer

```python
def create_model(activation_function, init):
  model = Sequential()
  model.add(Dense(64, input_dim = 16, kernel_initializer = init, activation = activation_function))
  model.add(Dropout(0.05))
  model.add(Dense(32, input_dim = 16, kernel_initializer=init, activation = activation_function))
  model.add(Dropout(0.05))
  model.add(Dense(26, activation = 'softmax'))
  adam = Adam(learning_rate = 0.003)
  model.compile(loss = 'categorical_crossentropy', optimizer = adam, metrics = ['accuracy'])
  return model
```

```python
model = KerasClassifier(build_fn = create_model, epochs = 20, batch_size = 8, verbose = 0, callbacks = [early_stop])
```

```python
param_dist = {
    'model__activation_function' : ['relu','tanh','leaky_relu'],
    'model__init' : ['uniform','normal','zero']
}
```

```python
grid_search2 = RandomizedSearchCV(estimator = model, param_distributions = param_dist,
                                  cv = KFold(), verbose = 0,n_iter = 10, n_jobs = -1)
```

```python
grid_result2 = grid_search2.fit(X_train, y_train)
```

```python
print('best %f using %s'%(grid_result2.best_score_, grid_result2.best_params_))
```

### Tuning of Hyperparameter: Number of neurons in activation layer

```python
def create_model(neuron1, neuron2):
  model = Sequential()
  model.add(Dense(neuron1, input_dim = 16, kernel_initializer='normal',activation = 'relu'))
  model.add(Dropout(0.1))
  model.add(Dense(neuron2, input_dim = 16, kernel_initializer='normal',activation = 'relu'))
  model.add(Dropout(0.1))
  model.add(Dense(26, activation = 'softmax'))
  adam = Adam(learning_rate = 0.001)
  model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'],optimizer = adam)
  return model
```

```python
model = KerasClassifier(build_fn = create_model,callbacks=[early_stop], epochs = 100, batch_size = 16, verbose = 0)
```

```python
param_dist = {
    'model__neuron1' :[16,32,64],
    'model__neuron2':[8,16,32]
}
```

```python
grid_search3 = RandomizedSearchCV(estimator = model,param_distributions = param_dist, cv = KFold(), verbose = 5, n_iter =20, n_jobs =-1)
```

```python
grid_result3 = grid_search3.fit(X_train, y_train)
```

Initial Model: A basic neural network with two hidden layers is defined using Keras.

Hyperparameter Tuning:

Batch Size & Epochs: Best result with batch size 8 and 50 epochs (accuracy 92.1%).

### Model with optimal parameters

```python
def create_model():
  model = Sequential()
  model.add(Dense(64, input_dim = 16, kernel_initializer='normal', activation = 'relu'))
  model.add(Dropout(0.1))
  model.add(Dense(32, input_dim = 16, kernel_initializer = 'normal', activation = 'relu'))
  model.add(Dropout(0.1))
  model.add(Dense(26, activation ='softmax'))

  adam = Adam(learning_rate = 0.001)
  model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = adam)
  return model
```

```python
model = KerasClassifier(build_fn = create_model, epochs = 50, batch_size = 8, verbose = 0, callbacks = [early_stop])
```

```python
print('Classification Report for test data')
print(classification_report(y_test.argmax(axis = 1), yhat_test.argmax(axis = 1)))
```
